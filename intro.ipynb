```python
'''
pip install pandas yfinance matplotlib
'''
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import matplotlib.pyplot as plt

start = '2016-01-01'
end = ''

df = yf.download(['AaPL', 'MSFT', 'TSLA'], start)
#print(tesla_df)

print(df.head())
print(f'\n{df.tail()}')

# Getting the Adj Close stock prices
df2 = df.loc[:, ['Adj Close']]
df2.rename(columns={'Adj Close':'adj_close'}, inplace=True)
#print(df2.head())

print(df2.adj_close.AAPL)

df2["simple_rtn"] = df2.adj_close.AAPL.pct_change()
df2["log_rtn"] = np.log(df2.adj_close.AAPL/df2.adj_close.AAPL.shift(1))
print(df2)

    
```
